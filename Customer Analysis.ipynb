```python
# trends in submissions
df_transcript['submitMonth'] = df_transcript['submitDate'].dt.to_period('M')
submission_trends = df_transcript.groupby('submitMonth').size()
```

```python
print(submission_trends)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
submission_trends.plot(kind='bar', color='skyblue')
plt.title('submission trends over time')
plt.xlabel('months')
plt.ylabel('number of submissions')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
# merging both tables
df_transactions['Phone']=df_transactions['Phone'].astype(str).str.replace('^237', '', regex=True)
df_transactions
```
